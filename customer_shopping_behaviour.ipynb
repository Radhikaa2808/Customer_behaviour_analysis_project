```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

The data is now cleaned from missing values using the right method.

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns = {'purchase_amount_(usd)': 'purchase_amount'})
```

We are changing different column names to 1 proper format so that we dont have to remember the syntax of each column name while performing ananlysis.

```python
df.columns
```

Will perform feature engineering to create useful columns from existing columns.

```python
#create a new column age_group 
labels = ['Young Adult', 'Adult', 'Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(20)
```

```python
# create another column purchase_frequency_days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
